Define evenly spaced grid points within spatial domain that's 2 units of length wide, i.e. $x_i \in (0, 2)$.

nx - number of grid points we want 

dx - distance between any pair of adjacent grid points

```python
nx = 41
dx = 2 / (nx - 1)
nt = 25
dt = .025
c = 1
```

Set up initial conditions as 

u = 2 in interval $0.5 \leq x \leq 1$ and $u=1$ everywhere else in $(0, 2)$

```python
u = np.ones(nx)
u[int(.5 /dx) : int(1 / dx + 1)] = 2
```

```python
pyplot.plot(np.linspace(0, 2, nx), u)
```

```python
# initialize temporary array
un = np.ones(nx)
for n in range(nt):
    # Copy existing values of u into un
    un = u.copy()
    for i in range(1, nx):
    # for i in range(nx):
        u[i] = un[i] - c * dt/dx * (un[i] - un[i - 1])


    
```
